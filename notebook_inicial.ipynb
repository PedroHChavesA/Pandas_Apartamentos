# Manipulando os dados

## Criando colunas numéricas

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados ['Condominio']
dados.head()
```

```python
dados["Valor_por_ano"] = (dados['Valor_por_mes'] * 12) + dados['IPTU']
dados.head()
```

## Criando colunas categóricas

```python
dados['Descricao'] = dados['Tipo'] + " Em " +dados['Bairro']
dados.head()
```

```python
# dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + dados['Quartos'] + ' quarto(s) ' + ' e ' + dados['Vagas'] + ' vaga(s) de garagem.'
                                        
# dados.head()
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + dados['Quartos'].astype(str) + ' quarto(s) ' + ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
                                        
dados.head()
```

```python
dados['Possui_Suite'] = dados['Suites'].apply(lambda x: 'Sim' if x > 0 else 'Nao')
dados.head()
```

```python
dados.to_csv('Dados_completos_DEV.csv', sep=';', index=False)
```
